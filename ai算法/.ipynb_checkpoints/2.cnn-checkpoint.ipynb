# 1.padding and conv卷积

![https://upload-images.jianshu.io/upload_images/15715517-d4da188fe047cd47.gif](https://upload-images.jianshu.io/upload_images/15715517-d4da188fe047cd47.gif)

m=3,padding=2,step=1

```python
# padding是四周都补0
def padding1(x,padding):
    zero=torch.zeros((x.size()[0],padding)).float()
    x1=torch.cat((x,zero),dim=1)
    zero=torch.zeros((x.size()[0],padding)).float()
    x1=torch.cat((zero,x1),dim=1)
    zero=torch.zeros((padding,x1.size()[1])).float()
    x1=torch.cat((x1,zero),dim=0)
    zero=torch.zeros((padding,x1.size()[1])).float()
    x1=torch.cat((zero,x1),dim=0)
    return x1

def conv(data,W_conv,in_channel,out_channel,kernal_size,stride,padding):
#padding=(kernel_size-stride)/2 时输出的图片大小与原图一样kernel_size卷积核的尺寸，stride步长，out_channel卷积核的数量-输出的图片数
#     feature map = (input_size - kernel_size + 2 * padding) / stride + 1 
#W[out_channel,in_channel,kernel_1,kernel_2]
#data[batch_size,in_channel,conv_w,conv_h]
# result[batch_size，out_channel,w,h]
    size=data.size()
    result=torch.zeros(size[0],out_channel,size[2],size[3])
    for z in range(size[0]):
        for d in range(size[1]):
            x=padding1(data[z,d].float(),padding)                
#             print(x.size())
            for i in range(x.size()[0]-kernal_size+1):
                for j in range(x.size()[1]-kernal_size+1):
                    for k in range(out_channel):                        
                        result[z,k,i,j]=torch.sum(torch.mul(W_conv[k,d,:,:],x[i:i+kernal_size,j:j+kernal_size]))
    return result
```

```python
# With square kernels and equal stride
# test conv
filters = torch.linspace(1,25*32,25*32).resize(32,1,5,5)#W[out_channel,in_channel,kernel_1,kernel_2]
inputs = torch.linspace(1,10*28*28,10*28*28).resize(10,1,28,28)#data[batch_num,in_channel,conv_w,conv_h]
print(filters.size(),inputs.size())
con=torch.nn.functional.conv2d(inputs, filters, padding=2)
print(con)
print(conv(inputs,filters,1,32,5,1,2))

```

```python
def max_pool(step,data):
    size=data.size()
    w=size[2]//step
    h=size[3]//step
    print(w,h)
    result=torch.zeros(size[0],size[1],w,h).float()
    for s in range(size[0]):
        for z in range(size[1]):
            for i in range(w):
                for j in range(h):
#                     print(s,w,i,j)
                    a=data[s,z,i*step:(i+1)*step,j*step:(j+1)*step]
#                     print(a)
#                     print(torch.max(a).size())
#                     print(torch.max(a))
                    result[s,z,i,j]=torch.max(a)
#                     print(result[s,w,i,j])
                
#                     print(torch.max(a))
    return result
            

print(h_conv1.size())
h_pool1=max_pool(2,h_conv1)
print(h_pool1.size())
```

# 3.全连接
[dropout](https://www.cnblogs.com/ocean1100/articles/9370557.html)  
以前：  
h_fc1=torch.nn.functional.relu(torch.mm(h_pool2_flat,W_fc1)+b_fc1)  
采用dropout后计算公式就变成了：  

 

上面公式中Bernoulli函数，是为了以概率p，随机生成一个0、1的向量。


```python
def fc(h_pool_flat,W_fc,b_fc):
    size=h_pool_flat.size()
#     h_pool_flat=h_pool.reshape(-1,in_size)
    h_pool_dropout=h_pool_flat
    h_fc=torch.mm(h_pool_dropout,W_fc)+b_fc
   
    return h_fc

def dropout1(h_fc,dropout):
    a=torch.cat((torch.linspace(0,0,int(h_fc.size()[1]*dropout)),torch.linspace(1,1,int(h_fc.size()[1]*(1-dropout)))),dim=0)    
    return h_fc*a[torch.randperm(a.size(0))]

h_fc1=torch.nn.functional.relu(fc(h_pool2.reshape(-1,7*7*64),W_fc1,b_fc1))
print(h_fc1)
h_fc1_drop=h_fc1*dropout1(h_fc1,0.5)
```

```python
logits=fc(h_fc1_drop,W_fc2,b_fc2)
print(logits)
```

```python
prediction=torch.nn.functional.softmax(logits)
print(prediction)
```

```python
accurancy=torch.sum(((torch.argmax(prediction,dim=1)-train_set.train_labels[0:100])==0).float())/prediction.size()[0]
```
