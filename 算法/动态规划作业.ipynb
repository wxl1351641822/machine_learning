# 1.最大字段

## 1.1时间复杂度O($ n^2 $)
## 1.2分治法
将序列两两分开，最后只得一值
每次递归返回其最大子序列
再两两合并
合并后的最大子序列，只能是返回值之一或者两个返回值与其中间段的加和。
其时间复杂度为O($ nlog_2(n)$)

```python
def max_sub(A,i,j):
    k=0
    if(i==j):
        k=A[i]
        print(i,i,k)
        return i,i,k
    else:        
        mid=int((i+j)/2)
#         print(i,mid,j)
        i1,j1,k1=max_sub(A,i,mid)
        i2,j2,k2=max_sub(A,mid+1,j)      
        
        k0=0
        for z in range(j1+1,i2):
            k0+=A[z]
        k=k1+k2+k0
        print(k1,k2,k0)
        if(k1>k2):
            besti=i1
            if(k>=k1):
                bestj=j2
            else:
                k=k1
                bestj=j1
        else:
            bestj=j2
            if(k>=k2):
                besti=i1
            else:
                besti=i2
                k=k2    
        
        print(i,mid,j,besti,bestj,k)
        return besti,bestj,k

def divideAndConquer(A,i,j):
    besti,bestj,sum=max_sub(A,i,j)
    if(sum<0):
        sum=0
    return besti,bestj,sum

A=[1,-1,-1,2,3,1,-1]
besti,bestj,sum=divideAndConquer(A,0,6)
print("最大子段为：（",besti,",",bestj,"）, sum =",sum)

    
```

## 1.3动态规划
将1.2的分治算法中所有计算过的部分均存入state数组中，当需要重复计算时，直接从数组中取用即可  
时间复杂度O($ nlog_2(n)$)

```python
def max_sub1(A,state,i,j):
    k=0
#     print(i,j)
    if(i==j):
        k=A[i]        
        state[i][i]=A[i]
        print(i,i,k)
        return i,i,k
    else:        
        mid=int((i+j)/2)
#         print(i,mid,j)
        i1,j1,k1=max_sub1(A,state,i,mid)
        i2,j2,k2=max_sub1(A,state,mid+1,j)      
        
        k0=0
        start=j1
        end=i2
        if(state[j1+1][mid]!=0):            
            k0+=state[start+1][mid]
            start=mid
        if(state[mid+1][i2-1]!=0):
            k0+=state[mid+1][i2-1]
            end=mid+1            
        for z in range(start+1,end):
            k0+=A[z]
        state[start+1][end-1]=k0
        k=k1+k2+ state[start+1][end-1]
#         print(k1,k2,state[start+1][end-1])
        if(k1>k2):
            besti=i1
            if(k>=k1):
                bestj=j2
            else:
                k=k1
                bestj=j1
        else:
            bestj=j2
            if(k>=k2):
                besti=i1
            else:
                besti=i2
                k=k2    
        state[besti][bestj]=k
        print(i,mid,j,besti,bestj,k)
        return besti,bestj,k

def dynamic(A,i,j):
    state=[]
    for _ in range(len(A)):
        state.append([0]*len(A))
    
    besti,bestj,sum=max_sub1(A,state,i,j)
    if(sum<0):
        sum=0
    return besti,bestj,sum

A=[1,-1,-1,2,3,1,-1]
besti,bestj,sum=dynamic(A,0,6)
print("最大子段为：（",besti,",",bestj,"）, sum =",sum)
```

将y的各项视作二进制编码，可得0-$ 2^6 $  
首先是枚举法

```python
def doub(n,a,b,result):
    z=1
    for j in range(len(a)):
        z=z*2
#         print(z)
        for i in range(len(result[0])):
#             print(z,i)
            if(i % z==0):
#                 print(0,i)
                result[0][i]+=b[j]
            else:
#                 print(1,i)
                result[1][i]+=a[j]
    print(result[0])
    print(result[1])
n=6
a=[2,5,7,10,5,2]
b=[3,8,4,11,3,4]
result=[[0]*2**6]*2
doub(n,a,b,result)
```

# 3.条件极值问题
$ x_i $取值[0,1,2]，使$ x_i=y_i $ +y<sub>n+i</sub>,$ y_i $属于[0,1]  
$ c_i $为商品价值  
$ a_i $为商品重量  
$ a_i = $a<sub>n+i</sub>  
$ c_i = $c<sub>n+i</sub>  
若g线性，则可转换为0/1背包问题

```python



def bag(n,v, c, w):
	n=n*2
	w=w*2
	v=v*2
	value = [[0 for j in range(c + 1)] for i in range(n + 1)]
	for i in range(1, n + 1):
		for j in range(1, c + 1):
			if j < w[i - 1]:
				value[i][j] = value[i - 1][j]
			else:
				value[i][j] = max(value[i - 1][j], value[i - 1][j - w[i - 1]] + v[i - 1])
		# 背包总容量够放当前物体，取最大价值
	return value
def show(n, c, w, value):
    n=n*2
    w=w*2
    
    print('最大值为:',value[n][c])
    x = [0 for i in range(n)]
    j = c
    for i in range(n, 0, -1):
        if value[i][j] > value[i - 1][j]:
            x[i - 1] = 1
            j -= w[i - 1]
    print('xi取值为:,',x)
    for i in range(int(n/2)):
        
        x[i]=x[i]+x[int(i+n/2)]
        print('x[', str(i+1), ']=',str(x[i]), end='\n')
if __name__=='__main__':
    n=4
    c=10
    w=[2,3,4,4]#ai
    v=[1,2,5,6]#ci
    
    #xi=yi+yi+n,取值【0,1,2】，yi取值【0,1】
    value = bag(n,v,c,w)    
    show(n, c, w, value)
# 输出
```

# 4.可靠性设计
由题目可得：  
$ max(\Pi g_i(m_i))=max(\Sigma g_i(m_i)) $  
$\Sigma_{i=1}^nm_ic_i<=c$
其中，每一级至少有一个设备。
$ m_i=x_i+1 x_i取值0-正无穷大 $  
若$ min(c_i)*d<=c-\Sigma c_i且min(c_i)*(d+1)<=c-\Sigma c_i $  
则为了满足$ \Sigma_{i=1}^nm_ic_i<=c ,x_i取值{0,1,2,…，d}$  
$ w_k^i第i个阶段增加一个第k级设备，且满足约束条件时， $  
$ w_k^i $= $ {log(i+1)}\over{log(i)} $  
则此问题可转化为有d*n个物品的0\/1背包问题（价值计算略有不同）
