## 3.1 图灵机
* 图灵机 ：一个具有存储载体的，按照具体指令可完成，向左或向右移动放置标记，抹去标记以及在计算终止时停机等4种基本操作的虚拟机器，可以看作是描述算法的语言 
    * DTM(确定性图灵机）
        * 组成：
            * 有线状态控制器 
            * 读写头 
            * 存储载体：一条双向的具有无限多个格的线性带 
        * 基本操作
            * 移动放置标记：向左、向右
            * 抹去标记
            * 停机
        * 需要规定的信息：
            * 字符取用集合：$ \Gamma $
                * 输入字符集合 $ \Sigma $
                * 空白符：$ b \in \Gamma $ \ $ \Sigma $
            * 状态集合：Q
                * 初始状态：$ q_0 $
                * 两个停机状态：$ q_Y ,q_N $
            * 转移函数$ \delta :(Q $ \ $ \{q_Y,q_N\})X\Gamma \rightarrow QX \Gamma X \{l,r\} $

        * 运行
            * 输入字符串x:放在1到|x|格上
            * 空白b:其他格
            * 读写头：1
            * 状态控制器：初始状态
            * 按转移函数一步步地运行
               * q=$ q_Y 或 q_N $，停止并分别回答是/否
               * 不是：则继续转移
                   * $ \delta(q,s)=(q',s',\Delta) $
                   * 读写头：擦除s,写上s'

# 4 NP类问题
## 4.1 无向图的团的问题
* 问题描述
    * 例：给定一个有n个顶点的无向图G=(V,E)及整数 k
    * 问：G是否包含一个具有k个顶点的完全子图（团）？
>团：完全图  
    * 实例-字符串表示
        * G-邻接矩阵
        * 长度 $ m=n^2+logk+1 $
        * CLIQUE={w#v|$ w,v \in \{0,1\}^*$,以w为邻接矩阵的图G有一个k个定点的团，v-k的二进制表示}
* 非确定性算法设计
    1. 分解---$ O(n^2) $
        * 输入字符串:w#v
        * 分解w#v
        * 计算：
            * n=$ |w|^{1/2} $
            * 用v表示k
        * 约束：输入不是w#v形式，则拒绝
    2. 猜想模块：V的子集V'---$ O(n) $
        * V':非确定性地选出来的
        * 用集合特征向量表示V'---A[1...n]
    3. 检验：确定性地检查V'的团的性质---$ O(k^2) $
        * V'是团---接受$ q_Y $
        * 不是---拒绝$ q_N $
    * 总用时：$ O(n^2+k^2) $

```python
###选择V的一个子集V'
import random
def Clique():
    j=0
    n=5
    k=2
    A=[0]*n
    for i in range(0,n):
        m=random.choice([0,1])
        if m==0:
            A[i]=0
        else:
            A[i]=1
            j=j+1
    if j!=k:
        return False
    return True

Clique()
```

## 4.3 多项式非确定性问题--->DTM
* 定理：
    * NP问题：$ \Pi $(每一个）
    * p(x):存在一个多项式---多项式时间内可以解决NP问题
    * 确定性算法：可在$ O(2^{P(n)})$时间内解决问题$ \Pi $
    * 每一个NP问题都有一个$ O(2^{P(n)})$的确定性算法可解决他
    * 证明：
        * 算法A：问题$ \Pi $的一个多项式时间的不确定性算法
            * 多项式时间q(n)
                * 为不失一般性，设q可在多项式时间内被估值
            * 不确定性算法
            * $ \Pi $的
            * 意义：
                * 对于长度为n的每个被接受的**输入**，
                * 必然存在字符集$ \Gamma $上长度至多为q(n)的某个**猜想字符串**
                * 使得算法A的检验阶段，在不多于q(n)步内回答“是”
        * 若$|\Gamma |=k $
            * 要考虑的可能的猜测最多为$ k^{q(n)} $---因为猜想字符串长q(n)
        * 对于一个长为n的给定输入
            * 通过应用算法A的确定性检验阶段到相应的$ k^{q(n)} $多个可能猜想字符串中的每一个，
            * 直到停止或运行q(n)步
            * 我们可以肯定地发现A是否接受该输入
        * 若A在该时间内接受--是
        * 否则--非
        * --->这是个确定性算法
            * 时间复杂度的上界：$ q(n)k^q(n) $
            * 等价于：$ O(2^{P(n)})$  



# 5 NP-C(NP完全问题)

* 多项式变换：
    * 映射：语言$ L_1 \subset \Sigma_1^* 到另一个语言 L_2 \subset \Sigma_2^* 的多项式变换---> 映射f:\Sigma_1^* \rightarrow \Sigma_2^*$

## 6.1 第k重子集问题
* 第k重子集问题
    * 例：已知整数$ c_1,c_2,...,c_n,k,L$；
    * 问：存在k个不同的子集$ s_1,s_2,...,s_k \subseteq \{1,2,...,n\}$，使得对于 $ i=1,...,k,有 \Sigma_{j \in S_i} c_j\geq L ?$
    * 用**划分问题**证明它是NP难问题
* 证明
    * 设$ S={c_1,c_2,...,c_n} $是划分问题的一般性例子 
       * 设我们已经有个算法A[S,K,L]可以用来解第k重子集问题，则可       
     * 设计出求解划分算问题的算法如下 ：
         * 若$ \Sigma_{i=1}^n c_i $为奇数-->否
         * 假定$ \Sigma_{i=1}^n c_i $为偶数
             * 令$ L=0.5*\Sigma_{i=1}^n c_i $
             * 算法A[S,K,L]为子程序
             * 用下列二分搜索技术确定重至少是L的不同子集的数目$ K^* $---n次调用可找到
                 1. 令$ K_{min}=0,K_{max}=2^n $
                 2. 若$ K_{max}-K_{min}=1,置K^*=K_{min},且停机 $
                 3. 令$ K=(K_{min}+K_{max})/2 $,并调用算法A[S,K,L]
                 4. 若回答是，则令$ K_{min}=K $-->转2 (这个子集至少要这么大）
                    * 否，则$ K_{max}=K $-->2 （这个子集还不够大）
             * 再调用一次A[S,K\*,L+1]可得到结果
                 * 若结果为“是”---S中所有重至少为L的子集也必然满足其重至少为L+1
                     * ==>没有重为L的子集
                     * ==>对划分问题答案为否
                 * 若结果为“否”---存在S的一个子集，重数为L
                     * 对划分问题回答为是
             * 有上述迭代问题容易看出 
                 * 若假设每次调用算法A只需要单位时间 
                 * 总用时：多项式时间
                 * ==>则以上给出了求解划分问题的一个多项式时间算法 
                 * 也就是说我们证明了划分问题，可多项式时间规约到第k重子集问题
                 * 是NP难问题


## 6.2 TSP问题
* 旅行商问题不是NP的
    * 因为对其解的任意猜想，要验证他是否是最优，需要同所有其他的环游比较，
    * 这样的环游会有指数多个，
    * 因而不可能在多项式时间内完成 
* 证明 ：用哈密顿（Hamiton）回路问题（NPC问题）证明旅行商问题是NP难的 
    * 对图G构造旅行商问题：
        * 图G=(V,E)
        * |V|=n
        
        * 耗费时间：多项式时间
    * 图G有hamiton回路<==>上述旅行商问题有解，且解的长度为n
    * 所以是np困难的

**哈密顿回路**---过且仅过所有顶点一次的圈（每个点度为2）


# 7 总结
## 7.1 证明
* 证明$ \Pi $是NPC的
    * 找到一个NPC的$ \Pi $‘
    * 将$ \Pi $’多项式时间**变换**到$ \Pi $
    * 因为所有NP问题-->$ \Pi $'-->$ \Pi $（都是多想是时间变换）
* 证明$ \Pi $是NP难
    * 找到一个NPC的$ \Pi $‘
    * 将$ \Pi $’多项式时间**图灵**归约到$ \Pi $
    * 因为若$ \Pi $有多项式时间算法--->$ \Pi $'也有，反之不成立--->$ \Pi $至少和$ \Pi $'一样难

## 7.2 分类
* co-NP：$ \Pi \in co-NP $<===>
    * 其补问题是NP问题
    * 即，co_NP问题的否定回答可在多项式时间内被验证 
* 如果NP=coNP
    * NP类在补运算下是封闭的 ,
        * $ \Pi \in NP $<==>$ \Pi 的补运算 \in NP $
    * 一个np问题无解的回答，将有一个简短的certificate
