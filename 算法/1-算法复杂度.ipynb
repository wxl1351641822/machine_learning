# 1 算法

* 计算机算法的特征：
* 数据输入与输出；
    * 确定性： 算法的每一步计算（包括判断）必须要有确切的定义，即每一步计算动作必须是清楚的，**无二义性**。
    * 可实现性：每种计算至少在原理上能由人用纸和笔在**有限的时间**内完成。
    * 有穷性： 一个算法总是在执行了**有穷步之后终止**

**统计执行步数**
* 程序步：独立程序片段，执行时间与实例特征无关；
* 执行语句：一个表意完整的程序语句；
* s/e : 每个语句中包含的执行程序步数；
* 频率：语句出现的次数；  
* 总执行步数＝$ \Sigma \omega_i k_i$

eg  


| 语句 | s/e | 频率 | 执行步数 | 
| --- | --- | --- | --- | 
| Void Addm(T \*\*a, ··· ) | 0 | 0 | 0 | 
| { | 0 | 0 | 0 |
| for(int i=0; i<rows; i++) | 1 | rows+1 | rows+1 |
| for(int j=0; j<cols; j++) | 1 | rows\*(cols+1) | rows\*cols+rows | 
| c[i][j]=a[i][j]+b[i][j]; | 1 | rows\*cols | rows\*cols | 
| } | 0 | 0 | 0 | 



执行步数总计
2\*rows\*cols+2\*rows+1

```python
# 使用计数器矩阵加法与执行步数统计--伪代码
template<class T>
void Addm(T **a, T **b, T **c, int rows, int cols)
{#//矩阵a和b相加得到矩阵c 
    for(int i=0; i<rows; i++){ 
        count++; #//对应于上一条for语句 (共执行了rows步)
        for(int j=0; j<cols; j++){
            count++; #对应于上一条for语句 (共执行了rows×cols步)
            c[i][j]=a[i][j]+b[i][j]; 
            count++;# //对应于赋值语句 (共执行了rows×cols步) 
        }
        count++;#//对应j的最后一次for循环 (共执行了rows步) •
    }
    count++#//对应i的最后一次for循环 (共执行了1步) • 
}
#此结果同上
```

## 渐进函数

* 计算复杂性：反映程序运行时间和占用空间随实例特征(如：输入规模n)变化的情况
* 复杂性函数：T(n)或S(n)---非负，递增
* 渐进函数：g(n)  
    * $ \displaystyle \lim_{n \rightarrow \infty} (f(n)-g(n))/f(n)=0 $
    * 同阶：$ \displaystyle \lim_{n \rightarrow \infty} f(n)/g(n)=c(常数） $
* f(n)的主项P(n)：* $ \displaystyle \lim_{n \rightarrow \infty} p_i(n)/p(n)=0 $
    * $ f(n)=p(n)+p_1(n)+p_2(n)+...+p_k(n) $
* 选g(n)=p(n):选主项为渐进函数

| - | 渐进上界f(n)=O(g(n)) | 渐进下界 $ f (n) = \Omega(g(n)) $  | 渐进同阶$ f (n) = \Theta(g(n)) $  |
| --- | --- | --- | --- |
| 定义 | $ f(n) \leq c·g(n) $ | $ f(n) \geq c·g(n) $ | - | 
| 极限定义 | $ \displaystyle \lim_{n \rightarrow \infty} f(n)/g(n) \leq c(常数） $ | $ \displaystyle \lim_{n \rightarrow \infty} f(n)/g(n) \geq c(常数） $  |  $ \displaystyle \lim_{n \rightarrow \infty} f(n)/g(n) = c(常数） $  |
| 注意 | 1. 记号f(n)=O(g(n))不能写成g(n)=O(f(n));2. 最后一个是松散的上界，最好给出最小渐进上界。 | 容易发现一个渐近下界，但人们更关注的是找到最大的渐进下界。 | 一般函数主项即是该函数的渐进同阶 |
