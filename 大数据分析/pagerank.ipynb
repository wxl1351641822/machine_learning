# 1.pagerank的M趋近于特征向量

```python
import numpy as np
```

```python
A=np.array([[0.5,0.5,0],[0.5,0,1],[0,0.5,0]])
U,V=np.linalg.eig(A)
print(U)
print(V)
```

```python
print("最大特征值1对应的列向量为")
print(V[:,2])
```

```python
print(V[:,2].reshape(3,1))
```

```python
print(np.matmul(A,V[:,2].reshape(3,1)))
```

```python
###pagerank迭代
r=np.array([1/3,1/3,1/3]).reshape(3,1)
for i in range(100):
    r=np.matmul(A,r)
    print(i,r.reshape(1,3))
```

```python
print(np.matmul(A,r))
```

# 2.正常pagerank的求法

```python
import networkx as nx
import matplotlib.pyplot as plt
import random
```

```python
Graph=nx.DiGraph()
Graph.add_nodes_from(range(0,100))
for i in range(800):
    j=random.randint(0,100)
    k=random.randint(0,100)
    Graph.add_edge(k,j)
```

```python
nx.draw(Graph,with_labels=True)
plt.show()
```

```python
pr=nx.pagerank(Graph,max_iter=100,alpha=0.01)
print(pr)
```

```python
print(pr)
```

```python
import operator
print(max(pr.items(),key=operator.itemgetter(1)))
```

```python
sum(pr.values())
```
