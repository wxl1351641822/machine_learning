# 1.pagerank的M趋近于特征向量

```python
import numpy as np
```

```python
A=np.array([[0.5,0.5,0],[0.5,0,1],[0,0.5,0]])
U,V=np.linalg.eig(A)
print(U)
print(V)
```

```python
print("最大特征值1对应的列向量为")
print(V[:,2])
```

```python
print(V[:,2].reshape(3,1))
```

```python
print(np.matmul(A,V[:,2].reshape(3,1)))
```

```python
###pagerank迭代
r=np.array([1/3,1/3,1/3]).reshape(3,1)
for i in range(100):
    r=np.matmul(A,r)
    print(i,r.reshape(1,3))
```
