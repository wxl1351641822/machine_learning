## Fisher判别
d维度空间-->1维度
* 原来可分-->不可分：选某个可以分的方向投影
* $ y_n=w^Tx:$
    * w---决定方向
    * 找最佳投影方向--类内距离最小，类间距离最大
    * 准则函数：$ J_F(w)={{(m_1(y)-m_2(y))^2}\over {S_1^2(y)+S_1^2(y)}}={{w^TS_bw}\over {w^TS_ww}} $——使之最大
        * $ m_1(y)、m_2(y) $：y_n的均值---上面为类间距离
            * $ m_i(y)={{1}\over{N_i}}\Sigma{y \in \omega_i }y={{1}\over{N_i}}\Sigma{x \in \omega_i }w^Tx=w^T{{1}\over{N_i}}\Sigma{x \in \omega_i }x=w^Tm_i $
            * $ (m_1(y)-m_2(y))^2=(w^Tm_1-w^Tm_2)^2=(w^Tm_1-w^Tm_2)(m_1^Tw-m_2^Tm_2)=w^T(m_1-m_2)(m_1-m_2)^Tw=w^TS_bw $
        * $ S_1^2(y)，S_1^2(y) $ :y_n的散度（方差/协方差）——下面为类内距离
            * $ S_i^2(y)=\Sigma{y \in \omega_i }(y-m_i(y))^2=\Sigma{x \in \omega_i }(w^Tx-w^Tm_i)^2=w^T(\Sigma{x \in \omega_i }(x-m_i)(x-m_i)^T)w=w^TS_iw $
            * $ S_1^2(y)+S_1^2(y)=w^T(S_1+S_2)w=w^TS_ww $
            
            * 类内距离：$ S_w(y)^2={S_1^2(y)+S_1^2(y)} $
        * x的(d-维度）：
            * 均值$ m_1,m_2 $
            * 类内离散度矩阵Si(方差/协方差）
            * 总样本类内离散度矩阵Sw（对称的半正定矩阵）
            * 样本间离散度矩阵$ Sb=(m_1-m_2)(m_1-m_2)^T $
    * 利用准则函数求的最大w：
        * $ J_F(w) $取极大值时的w
        * 拉格朗日乘数法
            * 令$ {w^TS_ww}=c \neq 0 $
            * $ L(w,\lambda)= {{w^TS_bw}-\lambda ({w^TS_ww}-c)},\lambda为拉格朗日乘子 $
            * 求偏导：
                * $ \frac{\partial {L(w,\lambda)}}{\partial w}={{S_bw}-\lambda ({S_ww})}=0 $
                * $ {{S_bw^*}=\lambda ({S_ww^*})} ,w^*为最优解 $
                * $ {{S_w^{-1}S_bw^*}=\lambda ({w^*})}——w^*为特征值，求S_w^{-1}S_b的特征值 $
                    * $ S_b=(m_1-m_2)(m_1-m_2)^T $
                    * $ S_bw^*=(m_1-m_2)(m_1-m_2)^Tw^*=(m_1-m_2)R,R=(m_1-m_2)^Tw^*是标量 $
                    * $ {\lambda ({w^*})={S_w^{-1}S_bw^*}=S_w^{-1}(m_1-m_2)R} $
                    * $ {w^*}={{R}\over{\lambda}}S_w^{-1}(m_1-m_2) $
                    * 只需要方向，标量可略： $ {w^*}=S_w^{-1}(m_1-m_2) $
           
    

## 梯度法实例
* 准则函数：$ J(w,x)=|w^Tx|-w^Tx,一维 $
    * 错误：$ w^Tx<0, \dot J=-2,w(k+1)=w(k)+2C,矫正 $
    * 正确：$ w^Tx>0, \dot J=0,w(k+1)=w(k),不变 $
* 固定增量的逐次调整算法
    * 感知机算法是梯度法的特例：
    * 准则函数：$ J(w,x)=\frac{|w^Tx|-w^Tx}{2},一维 $
    * 错误：$ w^Tx<0, \dot J=-1,w(k+1)=w(k)+Cx_k,矫正 $
    * 正确：$ w^Tx>0, \dot J=0,w(k+1)=w(k),不变 $

## 最小平方误差LMSE,H-K,最小二乘法
* 求解Xw=b(约束，b全正）
    * X：N*(n+1),行>列，超定方程，无唯一确定解--->最小二乘法求近似解
    * $ J(w,x,b)=\frac{1}{2}\|Xw-b\|^2= \frac{1}{2}(Xw-b)(Xw-b)^T $
        * 梯度法：得到了w和b的关系：梯度得到b(k)-->w(k)-->e(k)
            * $ \frac{\partial {J(w,x,b)}}{\partial w}=X^T(Xw-b)=0 $ 
            * ==>$ X^TXw=X^Tb(X^TX是（n+1)(n+1)的方阵) $
            * ==>$ w=(X^TX)^{-1}X^Tb=X^\#b(X^\#=(X^TX)^{-1}X^T是伪逆)，w可以由b得到 $
        * $ b(k+1)=b(k)-C(\frac{\partial {J}}{\partial b})_{b=b(k)} $
            * b全正，J正，当C正时，应使
                * $ (\frac{\partial {J}}{\partial b})_{b=b(k)}\leq 0保证b为正  $
                * $ -2(\frac{\partial {J}}{\partial b})_{b=b(k)}=(Xw-b)+|Xw-b| $
                    * 当Xw(k)-b(k)>0:
                        * $ -(\frac{\partial {J}}{\partial b})_{b=b(k)}=Xw(k)-b(k) $
                    * 当Xw(k)-b(k)<0:
                        * $ -(\frac{\partial {J}}{\partial b})_{b=b(k)}=0 $
        * 所以：
        * 迭代公式：
            * $ e(k)=Xw(k)-b(k)$
            * $ b(k+1)=b(k)-C(e(k))+|e(k)|)_{b=b(k)，w=w(k)} $           
            * $ w(k+1)=X^\#b(k+1)=w(k)-C(e(k)+|e(k)|) $


## 9.势函数法（非线性）


* 一个样本一个势函数
* 计算积累位势函数K(x)
    * $ x_{k+1} \in \omega_1：K_k(x_{k+1})<0 $则分错了：修改K(x)
    * $ x_{k+1} \in \omega_1：K_k(x_{k+1})>0 $正确，不变

    * $ K(x,x_1),x_1 \in \omega_1 $
    * $ -K(x,x_1),x_1 \in \omega_2 $
* 迭代
    * 分错：
        * $ K_{k+1}(x)=K_{k}(x)+K(x,x_{k+1}),x_{k+1} \in \omega_1 $
        * $ K_{k+1}(x)=K_{k}(x)-K(x,x_{k+1}),x_{k+1} \in \omega_2 $
    * 正确：
        * 不变
* $ x=(x_1,x_2)^T,x_k=(x_{k_1}，x_{k_2})^T $
* 如何选择势函数
    1. K(x,xk)=K(xk,x)<==>当且仅当x=xk时最大
    2. x距离xk越远越小，无穷远则趋于0
    3. 光滑的，x距离xk越远越小
